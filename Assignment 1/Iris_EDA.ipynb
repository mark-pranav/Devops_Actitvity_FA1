```python
Q1 = np.percentile(data['SepalWidthCm'], 25)
Q3 = np.percentile(data['SepalWidthCm'], 75)
IQR = Q3 - Q1
print(IQR)

ll = Q1 - (1.5 * IQR)
ul = Q3 + (1.5 * IQR)

upper_outliers = data[data['SepalWidthCm'] > ul].index.tolist()
lower_outliers = data[data['SepalWidthCm'] < ll].index.tolist()
bad_indices = list(set(upper_outliers + lower_outliers))
drop = True

if drop:
    data.drop(bad_indices, inplace = True, errors = 'ignore')
    print('Outliers from {} is removed '.format(data['SepalWidthCm']))
```
