```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("C:/Users/gaonk/OneDrive/Desktop/ML/MLPractice/ML Dataset/ML Dataset/Country-data.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
print(data.isna().sum())
```

```python
data = data.drop('country', axis=1)
print(data.head())
```

```python
print(data.columns)
```

```python
X = data.iloc[:,[2,6]].values
print(X)
```

```python
from sklearn.cluster import KMeans
list1 = []

for i in range(1,11):
    Kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    Kmeans.fit(X)
    list1.append(Kmeans.inertia_)
    print(list1)
plt.plot(range(1,11), list1)
plt.title('Elbow Method Graph')
plt.xlabel('No. of clusters')
plt.ylabel('List1')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
print(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1])
print(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1])
print(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1])
print(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1])
print(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1])
```

```python
y_kmeans = kmeans.fit(X)
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
labels = y_kmeans.labels_
print(labels)
metrics.silhouette_score(X, labels, metric = 'euclidean')
```

```python
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```

```python
plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], s=100, color='red', label='Cluster1')
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], s=100, color='blue', label='Cluster2')
plt.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=100, color='orange', label='Cluster3')
plt.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], s=100, color='black', label='Cluster4')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
dendro = shc.dendrogram(shc.linkage(X, method="ward"))
plt.title('Dendrogram Plot')
plt.xlabel('Euclidian Distances')
plt.ylabel('Customers')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, linkage='ward')
y_kmeans = hc.fit(X)
labels = y_kmeans.labels_
metrics.silhouette_score(X, labels, metric = 'euclidean')
```

```python
y_kmeans = hc.fit_predict(X)
```
